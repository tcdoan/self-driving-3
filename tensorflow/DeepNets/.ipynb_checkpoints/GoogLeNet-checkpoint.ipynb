# GoogLeNet

Original GoogLeNet/Inception paper
- https://arxiv.org/pdf/1409.4842.pdf


Inception is also one of the models included in Keras Applications
- https://keras.io/api/applications/inceptionv3/

Perform the necessary pre-processing steps.
- While the original Inception model used a 224x224 input like VGG
- InceptionV3 actually uses a 299x299 input.

```python
from keras.preprocessing import image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

from keras.applications.inception_v3 import preprocess_input, InceptionV3, decode_predictions

img_path = 'images/zebra.jpg'
img = image.load_img(img_path, target_size=(299, 299))

x = image.img_to_array(img)
print(x.shape)

x = np.expand_dims(x, axis=0)
print(x.shape)

x = preprocess_input(x)
print(x.shape)

plt.imshow(img)

# Load the pre-trained model
model = InceptionV3(weights='imagenet', include_top=True)

# Perform inference on our pre-processed image
predictions = model.predict(x)

# Check the top 3 predictions of the model
print('Predicted:', decode_predictions(predictions, top=3)[0])
```
