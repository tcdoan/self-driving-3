# ResNet

- Paper: https://arxiv.org/pdf/1512.03385.pdf
- ResNet is also a model included in Keras Applications, under ResNet50.
  - https://keras.io/api/applications/resnet/#resnet50-function

- Need to do ImageNet-related pre-processing if you want to use the pre-trained weights for it. 

- ResNet50 has a 224x224 input size.

```python
from keras.preprocessing import image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

from keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions

img_path = 'images/dog_labrador.jpg'
img = image.load_img(img_path, target_size=(224, 224))

x = image.img_to_array(img)
print(x.shape)

x = np.expand_dims(x, axis=0)
print(x.shape)

x = preprocess_input(x)
print(x.shape)

plt.imshow(img)

# Load the pre-trained model
model = ResNet50(weights='imagenet', include_top=True)

# # Perform inference on our pre-processed image
predictions = model.predict(x)

# Check the top 3 predictions of the model
print('Predicted:', decode_predictions(predictions, top=3)[0])
```
