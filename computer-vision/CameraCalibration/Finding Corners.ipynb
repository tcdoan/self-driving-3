### Use findChessboardCorners() and drawChessboardCorners()  functions 
- #### To automatically find and draw corners in an image of a chessboard pattern.

### To learn more about both of those functions:
- ##### [ cv2.findChessboardCorners()](https://docs.opencv.org/2.4/modules/calib3d/doc/camera_calibration_and_3d_reconstruction.html#cv2.findChessboardCorners)
- ##### [cv2.drawChessboardCorners()](http://docs.opencv.org/2.4/modules/calib3d/doc/camera_calibration_and_3d_reconstruction.html#cv2.drawChessboardCorners) 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# prepare object points
nx = 8
ny = 6

# Make a list of calibration images
fname = 'calibration_test.png'
img = cv2.imread(fname)
plt.figure(figsize=(12,10))
plt.imshow(img)
```

```python
objPoints = []    
imgPoints = []

objP = np.zeros((8*6, 3), np.float32)
objP[:,:2] = np.mgrid[0:8, 0:6].T.reshape(-1,2)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, corners = cv2.findChessboardCorners(gray, (8,6), None)

if ret:
    objPoints.append(objP)
    imgPoints.append(corners)    
    cv2.drawChessboardCorners(img, (8,6), corners, ret)
    plt.figure(figsize=(12,10))
    plt.imshow(img)
    
```
